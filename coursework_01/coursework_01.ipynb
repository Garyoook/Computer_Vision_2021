### 2.2 Implement a function that generates a 2D Gaussian filter given the parameter $\sigma$. (10 points)

```python
# Design the Gaussian filter
def gaussian_filter_2d(sigma):
    # sigma: the parameter sigma in the Gaussian kernel (unit: pixel)
    #
    # return: a 2D array for the Gaussian kernel
    k = 3     # define the k default value as 3
    shape = (2*k*sigma, 2*k*sigma)
    x, y = [int(edge/2) for edge in shape]
    grid = np.array([[((i**2+j**2)/(2.0*sigma**2)) for i in range(-x, x+1)] for j in range(-y, y+1)])
    g_filter = np.exp(-grid)/(2*np.pi*sigma**2)
    g_filter /= np.sum(g_filter) # apply a normalisation here (so the sum of all values in the filter grid is 1)
                                 # to make the filtered output more stable (not too bright nor too dark)
    h = g_filter           
    return h

# Visualise the Gaussian filter when sigma = 3 pixel (provided)
sigma = 3
h = gaussian_filter_2d(sigma)
print('h: ')
print(h)
plt.imshow(h)
```

### 2.3 Perform Gaussian smoothing ($\sigma$ = 3 pixels), followed by Sobel filtering, show the gradient magnitude image. (7 points)

```python
# Perform Gaussian smoothing before Sobel filtering
g_filter = gaussian_filter_2d(3)

image_g_blurred = scipy.signal.convolve2d(image_noisy, g_filter, mode='same')

# Sobel filtering

sobel_x = scipy.signal.convolve2d(image_g_blurred, h_sobel_x, mode='same')
sobel_y = scipy.signal.convolve2d(image_g_blurred, h_sobel_y, mode='same')

# Calculate the gradient magnitude

sobel_mag_g = np.hypot(sobel_x, sobel_y)

# Display the magnitude image (provided)
plt.imshow(sobel_mag_g, cmap='gray', vmin=0, vmax=100)
plt.gcf().set_size_inches(10, 8)
```

### 2.5 Implement a function that generates a 1D Gaussian filter given the parameter $\sigma$. Generate 1D Gaussian filters along x-axis and y-axis respectively. (10 points)

```python
# Design the Gaussian filter
def gaussian_filter_1d(sigma):
    # sigma: the parameter sigma in the Gaussian kernel (unit: pixel)
    #
    # return: a 1D array for the Gaussian kernel
    k = 3       # define the k default value as 3
    sz = 2*k*sigma
    x = int(sz/2)
    grid = np.array([[((i**2)/(2.0*sigma**2)) for i in range(-x, x+1)]])
    g_filter = np.exp(-grid)/(np.sqrt(2*np.pi)*sigma)
    g_filter /= np.sum(g_filter) # apply a normalisation here (so the sum of all values in the filter grid is 1)
                                 # to make the filtered output more stable (not too bright nor too dark)
    h = g_filter
    return h

# sigma = 7 pixel (provided)
sigma = 7

# The Gaussian filter along x-axis. Its shape is (1, sz).
row = gaussian_filter_1d(sigma)[0]
h_x = []
for i in range(len(row)):
    h_x.append([row[i]])

# The Gaussian filter along y-axis. Its shape is (sz, 1).
h_y = gaussian_filter_1d(sigma)

print('h_x: {}'.format(h_x))
print('h_y: {}'.format(h_y))

# Visualise the filters (provided)
plt.subplot(1, 2, 1)
plt.imshow(h_x)
plt.subplot(1, 2, 2)
plt.imshow(h_y)
```

### 2.6 Perform Gaussian smoothing (sigma = 7 pixels) using two separable filters and evaluate the computational time for separable Gaussian filtering. After that, perform Sobel filtering and show results. (9 points)

```python
# Perform separable Gaussian smoothing and count time

start_time = time.clock_gettime(0)
image_g_blurred_x = scipy.signal.convolve2d(image_noisy, h_x, mode='same')
image_g_blurred_xy = scipy.signal.convolve2d(image_g_blurred_x, h_y, mode='same')
# print(image_g_blurred_xy)
end_time = time.clock_gettime(0)
time_spent_sep = end_time - start_time
print('time spent: {}s'.format(time_spent_sep))

# Sobel filtering

sobel_x = scipy.signal.convolve2d(image_g_blurred_xy, h_sobel_x, mode='same')
sobel_y = scipy.signal.convolve2d(image_g_blurred_xy, h_sobel_y, mode='same')

# Calculate the gradient magnitude

sobel_mag = np.hypot(sobel_x, sobel_y)

# Display the magnitude image (provided)
plt.imshow(sobel_mag, cmap='gray', vmin=0, vmax=100)
plt.gcf().set_size_inches(10, 8)
```

### 2.7 Comment on the Gaussian + Sobel filtering results and the computational time. (10 points)

The two sobel filters we applied find the discontinuity of the image along the x-axis and y-axis, the we combine them and calculate the magnitude to get the primary image with information about edges in the input image. Then we apply Gaussian filter to the output image from sobel filters in order to supress the noise, as we can see from 2.3 and 2.4, the more powerful Gaussian filter we use (larger sigma), the less noise we have on the output image. yet at the same time the output image get more blurry as we increase the intensity of Gaussian filtering.

As for the computational time, the separable Gaussian filter using an 1*7 kernel and a 7*1 kernel shows a significantly faster speed (complete in ~0.87s) than the Gaussian filter using a single 7*7 kernel (complete in ~3.6s). This is because by using separable kernels, we reduce the number of multiplications and summations for each pixel on the image and hence reduce the computational complexity from O(N^2K^2) (p.s. N^2K^2 multiplications and N^2(K^2 - 1) summations) to O(N^2K) (p.s. 2N^2K multiplications and 2N^2(K - 1) summations) where N is the number of pixels on the image and K is the size of kernel. The larger image we apply the filter on, the more obvious optimisation we can oberve in computation time.
